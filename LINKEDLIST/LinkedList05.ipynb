### 16 Intersection of 2 LinkedList
- https://leetcode.com/problems/intersection-of-two-linked-lists/description/

```python
### Using the hashmaping technique 
def getIntersectionNode(headA,headB):
        setA = set()
        temp1 = headA 

        while temp1:
            setA.add(temp1)
            temp1 = temp1.next
        
        temp2 = headB
        while temp2:
            if temp2 in setA:
                return temp2
            temp2 = temp2.next
        return None
```

- Time Complexcity: O(m+n)
- Space Complexcity: O(max(m,n))

```python
# Using the Length difference alignment technique
def getIntersectionNode(headA, headB):
        # To get the count of Linked List A, B
        count1 = 0
        count2 = 0
        temp1 = headA
        temp2 = headB 

        while temp1: # T O(n)
            count1 += 1
            temp1 = temp1.next

        while temp2: # T O(m)
            count2 += 1
            temp2 = temp2.next
        
        # T O(d)
        if count1 > count2:
            temp1 = headA 
            for i in range(abs(count1-count2)):
                temp1 = temp1.next
            temp2 = headB
        else:
            temp2 = headB 
            for i in range(abs(count1-count2)):
                temp2 = temp2.next
            temp1 = headA
        
        # Let's compare the nodes 
        while temp1 and temp2: # T O(max(m,n) - d)
            if temp1 == temp2:
                return temp1
            temp1 = temp1.next
            temp2 = temp2.next
        return None
```

- Time Complexcity: O(m+n+max(m,n))
- Space Complexcity: O(1)

```python
# Switching heads Technique
def getIntersectionNode(headA, headB):
        # Switching Head Technique 
        if headA is None or headB is None:
            return None
        tempA = headA
        tempB = headB

        while tempA != tempB:
            tempA = tempA.next if tempA else headB
            tempB = tempB.next if tempB else headA
        return tempA
```

- Time Complexcity: O(m+n)
- Space Complexcity: O(1)

### 17 DETECT A CYCLE IN LINKED LIST
- https://leetcode.com/problems/linked-list-cycle/description/

```python
# Using the hashmap
class Solution:
    def hasCycle(head):
        hash_map = set()
        temp = head

        while temp:
            if temp not in hash_map:
                hash_map.add(temp)
            else:
                return True
            temp = temp.next
        return False
```

- Time complexcity : O(n)
- Space complecity : O(n)

```python
# using the floyd's cycle detection algorithm
class Solution:
    def hasCycle(head):
        fast = head
        slow = head 

        while fast and fast.next:
            slow = slow.next 
            fast = fast.next.next 
            if slow == fast:
                return True 
        return False
```

- Time complexcity : O(m + n)
- Space Complexcity : O(1)

### 18 Remove Duplicates from Sorted List II
- https://leetcode.com/problems/remove-duplicates-from-sorted-list-ii/description/?envType=problem-list-v2&envId=linked-list

```python
class Solution:
    def deleteDuplicates(self, head):
        count_numbers = dict()
        temp = head
        dummy_list = ListNode(val='dummy')
        dummy = dummy_list

        while temp:
            if temp.val in count_numbers:
                count_numbers[temp.val] += 1
            else:
                count_numbers[temp.val] = 1
            temp = temp.next
        
        for key, value in count_numbers.items():
            if value == 1:
                dummy.next = ListNode(val=key)
                dummy = dummy.next
        return dummy_list.next
```

###  20 Linked List Cycle II
- https://leetcode.com/problems/linked-list-cycle-ii/description/?envType=problem-list-v2&envId=linked-list

```python
class Solution:
    def detectCycle(self, head):
        slow = head
        fast = head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                fast = head
                while fast != slow:
                    slow = slow.next
                    fast = fast.next
                return slow
        return None
```
