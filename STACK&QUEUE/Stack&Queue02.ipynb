### 6. MinStack 
- https://leetcode.com/problems/min-stack/submissions/1985863043/

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val: int) -> None:
        self.stack.append(val)

        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self) -> None:
        if self.stack[-1] == self.min_stack[-1]:
            self.min_stack.pop()
        self.stack.pop()

    def top(self) -> int:
        return self.stack[-1] if self.stack else -1

    def getMin(self) -> int:
        return self.min_stack[-1] if self.min_stack else -1


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

### 7. Next Greater Element 1
- https://leetcode.com/problems/next-greater-element-i/

```python
class Solution:
    def nextGreaterElement(self, nums1, nums2):
        greater_element = dict()
        stack = []

        for num in nums2:
            while stack and num > stack[-1]:
                smaller = stack.pop() 
                greater_element[smaller] = num 
            stack.append(num)
        
        while stack:
            greater_element[stack.pop()] = -1 
        
        return [greater_element[num] for num in nums1]
```

### 8. Next Greater Element II
- https://leetcode.com/problems/next-greater-element-ii/description/

```python
class Solution:
    def nextGreaterElements(self, nums):
        n = len(nums)
        stack = []
        res = [-1] * n

        for i in range(2*n):
            while stack and nums[i%n] > nums[stack[-1]]:
                min_index = stack.pop()
                res[min_index] = nums[i%n]
            if i < n:
                stack.append(i)
        return res
```

### 9. Previous Smaller Element
- https://www.geeksforgeeks.org/problems/previous-smaller-element/1

```python
class Solution:
	def prevSmaller(self, arr):
		# code here
		n = len(arr)
		previous_smaller = [-1] * n
		
		for i in range(n):
		    for j in range(i-1, -1, -1):
		        if arr[j] < arr[i]:
		            previous_smaller[i] = arr[j]
		            break
		return previous_smaller
```

### 10. Trapping Rain water 
- https://leetcode.com/problems/trapping-rain-water/description/

```python
class Solution:
    # Function: for the prefix sum 
    def prefix_max(self, nums):
        prefixMax = [0] * len(nums)
        prefixMax[0] = nums[0]

        for i in range(1,len(nums)):
            prefixMax[i] = max(nums[i], prefixMax[i-1])
        return prefixMax
    
    # FUnction for the suffix max 
    def suffix_max(self, nums):
        suffixMax = [0] * len(nums)
        suffixMax[-1] = nums[-1]

        for i in range(len(nums)-2, -1, -1):
            suffixMax[i] = max(nums[i], suffixMax[i+1])
        return suffixMax


    def trap(self, height):
        n = len(height)
        prefixMax = self.prefix_max(height)
        suffixMax = self.suffix_max(height)
        total = 0

        for i in range(n):
            if height[i] < prefixMax[i] and height[i] < suffixMax[i]:
                total += (min(prefixMax[i], suffixMax[i]) - height[i])
        return total
```
