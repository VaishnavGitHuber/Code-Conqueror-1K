### 1. Valid Paranthesis 
- https://leetcode.com/problems/valid-parentheses/description/

```python
class Solution:
    def isValid(self, s: str) -> bool:
        mapping = {')': '(', '}':'{', ']':'['}
        stack = []

        for char in s:
            if char in mapping:
                top = stack.pop() if stack else '#'
                if top != mapping[char]:
                    return False
            else:
                stack.append(char)
        return True if not stack else False
```

### 2. Infix to Postfix 
- https://www.geeksforgeeks.org/problems/infix-to-postfix-1587115620/1

```python
class Solution:
    def infixtoPostfix(self, s):
        #code here
        stack = []
        result = ""
        precedence = {'^':3, '*':2, '/':2, '+':1, '-':1}
        
        for char in s:
            if char.isalnum():
                result += char
            
            elif char == '(':
                stack.append(char)
            
            elif char == ')':
                while stack and stack[-1] != '(':
                    result += stack.pop()
                stack.pop()
            
            else:
                # handling the right asociative and left asociative
                while stack and stack[-1] != '(' and (char != '^' and (precedence[char] <= precedence[stack[-1]]) or (char == '^' and precedence[char] < precedence[stack[-1]])):
                    result += stack.pop()
                stack.append(char)
            
        while stack:
            result += stack.pop()
        
        return result
```

### 3. Infix to Prefix
- https://www.geeksforgeeks.org/problems/infix-to-prefix-notation/1

```python
class Solution:
    def infixToPrefix(self, s):
        #code here
        reverse_string = reversed(s)
        precedence = {'^':3, '*':2, "/":2, "+":1, "-":1}
        stack = []
        result = []
        
        for char in reverse_string:
            if char.isalnum():
                result.append(char)
            
            elif char == ')':
                stack.append('(')
            
            elif char == '(':
                while stack and stack[-1] != '(':
                    result.append(stack.pop())
                stack.pop()
            
            else:
                while stack and stack[-1] != '(' and ((char != '^' and precedence[char] < precedence[stack[-1]]) or (char == '^' and precedence[char] <= precedence[stack[-1]])):
                    result.append(stack.pop())
                stack.append(char)
        while stack:
            result.append(stack.pop())
        
        return ''.join(reversed(result))
```

### 4. Postfix to Infix 
- https://www.geeksforgeeks.org/problems/postfix-to-infix-conversion/1

```python
#User function Template for python3
class Solution:
    def postToInfix(self, postfix):
        # Code here
        stack = []
        
        for char in postfix:
            if char.isalnum():
                stack.append(char)
            else:
                opr1 = stack.pop()
                opr2 = stack.pop()
                
                new_exp = '(' + opr2 + char + opr1 + ')'
                stack.append(new_exp)
        return stack[0]
```

### 5. Prefix to infix 
- https://www.geeksforgeeks.org/problems/prefix-to-infix-conversion/1

```python
#User function Template for python3
class Solution:
    def preToInfix(self, pre_exp):
        # Code here
        stack = []
        
        for char in pre_exp[::-1]:
            if char.isalnum():
                stack.append(char)
            else:
                opr1 = stack.pop()
                opr2 = stack.pop()
                
                new_exp = '(' + opr1 + char + opr2 + ')'
                
                stack.append(new_exp)
        return stack[0]
```
