### 16. Largest Rectangle in a histogram
- https://leetcode.com/problems/largest-rectangle-in-histogram/

```python
class Solution:
    def largestRectangleArea(self, heights):
        n = len(heights)
        stack = []
        previous_smaller = [-1] * n
        next_smaller = [n] * n
        maximum_area = 0
        
        # Previous smllaer
        for i in range(n):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            previous_smaller[i] = stack[-1] if stack else -1
            stack.append(i)
        
        stack = []
        # Next smaller
        for i in range(n-1, -1, -1):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            next_smaller[i] = stack[-1] if stack else n
            stack.append(i)
        
        for i in range(n):
            width = next_smaller[i] - previous_smaller[i] - 1
            area = heights[i] * width 
            maximum_area = max(maximum_area, area)
        return maximum_area

        
```

### 17. PREVIOUS GREATER ELEMENT
- https://www.geeksforgeeks.org/problems/previous-greater-element/1

```python
class Solution:
	def preGreaterEle(self, arr):
		# code here
		n = len(arr)
		stack = []
		previous_greater = [-1] * n 
		
		for i in range(n-1, -1, -1):
		    while stack and arr[i] > arr[stack[-1]]:
		        min_index = stack.pop()
		        previous_greater[min_index] = arr[i]
		    stack.append(i)
		return previous_greater
```

### 19. DAILY TEMPARATURE 
- https://leetcode.com/problems/daily-temperatures/

```python
class Solution:
    def dailyTemperatures(self, temperatures):
        n = len(temperatures)
        output = [0] * n
        stack = []

        for i in range(n):
            while stack and temperatures[i] > temperatures[stack[-1]]:
                min_index = stack.pop()
                output[min_index] = i - min_index
            stack.append(i)
        return output
```

### 20. MAXIMUL R
- https://leetcode.com/problems/maximal-rectangle/description/

```python
class Solution:
    def largestRectangleArea(self, heights):
        n = len(heights)
        previous_smaller = [-1] * n # only store index for width 
        max_area = 0
        stack = []

        for i in range(n):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            previous_smaller[i] = stack[-1] if stack else -1
            stack.append(i)
        
        stack = []
        next_smaller = [n] * n

        for i in range(n-1, -1, -1):
            while stack and heights[stack[-1]] >= heights[i]:
                stack.pop()
            next_smaller[i] = stack[-1] if stack else n
            stack.append(i)
        
        for i in range(n):
            width = next_smaller[i] - previous_smaller[i] - 1
            area = width * heights[i]
            max_area = max(max_area, area)
        return max_area

    def maximalRectangle(self, matrix):
        row, col = len(matrix), len(matrix[0])
        heights = [0] * col
        max_area = 0

        for i in range(row):
            for j in range(col):
                if matrix[i][j] == '1':
                    heights[j] += 1
                else:
                    heights[j] = 0
            max_area = max(max_area, self.largestRectangleArea(heights))
        return max_area
```
