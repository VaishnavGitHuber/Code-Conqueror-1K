### 1. Battleships in a board 
- https://leetcode.com/problems/battleships-in-a-board/description/
- No_island

```python
class Solution:
    def countBattleships(self, board):
        m, n = len(board), len(board[0])
        no_battleships = 0
        dirs = [(1, 0), (0, 1)]

        # DFS Algorithm 
        def bfs(row, col):
            queue = [(row, col)]

            while queue:
                row, col = queue.pop(0)

                for r, c in dirs:
                    if 0 <= row+r < m and 0 <= col+c < n and board[row+r][col+c] == 'X':
                        board[row+r][col+c] = '.'
                        queue.append((row+r, col+c))

        for i in range(m):
            for j in range(n):
                if board[i][j] == 'X':
                    bfs(i, j)
                    no_battleships += 1
        return no_battleships
```

### 2. Max Area of Island
- https://leetcode.com/problems/max-area-of-island/description/
- no_island

```python
class Solution:
    def maxAreaOfIsland(self, grid):
        m, n  = len(grid), len(grid[0])
        max_area = 0
        dirs = [(1, 0), (-1, 0), (0, 1), (0, -1)]

        def bfs(row, col):
            grid[row][col] = 0
            queue = [(row, col)]
            area = 1

            while queue:
                row, col = queue.pop(0)

                for r,c in dirs:
                    nr,nc = r + row, c + col

                    if 0 <= nr < m and 0 <= nc < n and grid[nr][nc] == 1:
                        grid[nr][nc] = 0
                        area += 1
                        queue.append((nr, nc))
            return area
            

        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    max_area = max(max_area, bfs(i,j))
        return max_area
```

### 3. Graph Valid Tree
- https://www.naukri.com/code360/problems/graph-valid-tree_1376618?leftPanelTabValue=SUBMISSION
- Cycle detection in undirected graph(no cycle, connected graphs)

```python
from os import *
from sys import *
from collections import *
from math import *

def checkgraph(edges, n, m):
    adj_list = [[] for _ in range(n)]

    for (u, v) in edges:
        adj_list[u].append(v)
        adj_list[v].append(u)
    
    visited = [False] * n

    queue = [(0, -1)]
    visited[0] = True

    while queue:
        node, parent = queue.pop(0)

        for neighbour in adj_list[node]:
            if not visited[neighbour]:
                queue.append((neighbour, node))
                visited[neighbour] = True
            else:
                if neighbour != parent:
                    return False # Not valid tree
    return False if sum(visited) != n else True
```

### 4. Island Perimeter 
- https://leetcode.com/problems/island-perimeter/

```python
class Solution:
    def islandPerimeter(self, grid):
        # If the land share it's border with another land, then it is not a part of border 
        ## Perimeter = 4 * (no of land) - 2 * (shared land) 
        m, n = len(grid), len(grid[0])
        perimeter = 0

        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    perimeter += 4

                    # To the right 
                    if j+1 <= n-1 and grid[i][j+1] == 1:
                        perimeter -= 2
                    
                    # To the down 
                    if i+1 <= m-1 and grid[i+1][j] == 1:
                        perimeter -= 2
        return perimeter
```

### 5. No of Closed Island
- https://leetcode.com/problems/number-of-closed-islands/

```python
class Solution:
    def closedIsland(self, grid):
        m, n = len(grid), len(grid[0])

        # DFS Algorithm 
        def dfs(row, col):
            if 0 <= row < m and 0 <= col < n and grid[row][col] == 0:
                grid[row][col] = 1 # water 

                dfs(row+1, col)
                dfs(row-1, col)
                dfs(row, col+1)
                dfs(row, col-1)
        
        # Remove the outer boundry connected islands 
        for i in range(m):
            for j in range(n):
                if i == 0 and grid[i][j] == 0:
                    dfs(i, j)
                if i == m-1 and grid[i][j] == 0:
                    dfs(i, j)
                if j == 0 and grid[i][j] == 0:
                    dfs(i, j)
                if j == n-1 and grid[i][j] == 0:
                    dfs(i, j)
        # No of islands
        no_closed_islands = 0

        for i in range(m):
            for j in range(n):
                if grid[i][j] == 0:
                    dfs(i, j)
                    no_closed_islands += 1
        return no_closed_islands
```
