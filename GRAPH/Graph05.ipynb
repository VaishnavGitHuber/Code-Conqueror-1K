### 21 Dijkstra's Algorithm(heap queue DS)
- https://www.geeksforgeeks.org/problems/implementing-dijkstra-set-1-adjacency-matrix/1

```python
import heapq

class Solution:
    # Returns shortest distances from src to all other vertices
    def dijkstra(self, V, edges, src):
        # Step 1: Make a graph datastructure 
        graph = [[] for _ in range(V)]
        
        for u,v,w in edges:
            graph[u].append((v,w))
            
            graph[v].append((u,w))
            
        # Step 2: Distance array
        INF = float('inf')
        dist = [INF] * V
        dist[src] = 0
        
        # Step 3: Min heap 
        pq = []
        heapq.heappush(pq, (0, src)) # (dist, node)
        
        # Step 4: Dijitras algorithm 
        while pq:
            curr_dist, u = heapq.heappop(pq)
            # SAFETY CHECK
            if curr_dist > dist[u]:
                continue
            
            for v,wt in graph[u]:
                if dist[u] + wt < dist[v]:
                    dist[v] = dist[u] + wt 
                    heapq.heappush(pq, (dist[v], v))
        return dist
```

### 22 Dijkstra's Algorithm(Set DS)
- https://www.geeksforgeeks.org/problems/implementing-dijkstra-set-1-adjacency-matrix/1

```python
class Solution:
    # Returns shortest distances from src to all other vertices
    def dijkstra(self, V, edges, src):
        # code here
        # Step 1 : Making the graph data structure
        graph = [[] for _ in range(V)]
        
        for u,v,w in edges:
            graph[u].append((v,w))
            graph[v].append((u,w))
        
        # Step 2: Distance array 
        INF = float('inf')
        dist = [INF] * V
        dist[src] = 0
        
        # Step 3: Set datastructure 
        st = set()
        st.add((0, src))
        
        # Step 4: Process the set datastructure 
        while st:
            curr_dist, u = min(st)
            st.remove((curr_dist, u))
            
            if curr_dist > dist[u]:
                continue
            
            for v, wt in graph[u]:
                if dist[u] + wt < dist[v]:
                    if (dist[v], v) in st:
                        st.remove((dist[v], v))
                        
                    dist[v] = dist[u] + wt
                    st.add((dist[v], v))
        return dist
```

### 23 BELLMAN FORD SHORTEST DISTANCE ALGORITHM
- https://www.geeksforgeeks.org/problems/distance-from-the-source-bellman-ford-algorithm/1

```python
#User function Template for python3

class Solution:
    def bellmanFord(self, V, edges, src):
        #Step1 : Distance array 
        INF = 10**8
        dist = [INF] * V 
        dist[src] = 0
        
        # Step 2: Relax the N-1 vetex
        for _ in range(V-1):
            for u,v,w in edges:
                if dist[u] != INF:
                    if dist[u] + w < dist[v]:
                        dist[v] = dist[u] + w
        
        # Step 3: Neigative cycle detection 
        for u,v,w in edges:
            if dist[u] != INF and  dist[u] + w < dist[v]:
                return [-1]
        
        
        return dist
                    
```

### 24 NEGATIVE WEIGHT CYCLE
- https://www.geeksforgeeks.org/problems/negative-weight-cycle3504/1

```python
#User function Template for Python3

def isNegativeWeightCycle(self, n, edges):
        #Step1 : Distance array 
        dist = [0] * n
        
        # Step 2: Relax the N-1 vertex
        for _ in range(n-1):
            for u,v,w in edges:
                if dist[u] + w < dist[v]:
                    dist[v] = dist[u] + w
        
        # Step 3: Negative cycle detection 
        for u,v,w in edges:
            if dist[u] + w < dist[v]:
                return 1
        
        
        return 0
```

### 25 Find if Path Exists in Graph
- https://leetcode.com/problems/find-if-path-exists-in-graph/description/

```python
class Solution:
    def validPath(self, n: int, edges, source, destination):
        visited = [False] * n 
        Graph = [[] for i in range(n)]

        for (u,v) in edges:
            Graph[u].append(v)
            Graph[v].append(u)

        queue = [source]
        visited[source] = True

        while queue:
            node = queue.pop(0)

            if node == destination:
                return True

            for neighbour in Graph[node]:
                if not visited[neighbour]:
                    queue.append(neighbour)
                    visited[neighbour] = True
                

        return False
```
