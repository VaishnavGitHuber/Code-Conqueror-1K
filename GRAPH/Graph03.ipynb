#### 11 Number of Distinct island
- https://www.geeksforgeeks.org/problems/number-of-distinct-islands/1

```python
#User function Template for python3

import sys
from typing import List
sys.setrecursionlimit(10**8)

class Solution:
    def countDistinctIslands(self, grid : List[List[int]]) -> int:
        # take the boundry of the islands 
        m,n = len(grid), len(grid[0])
        visited = [[False] * n for i in range(m)]
        shapes = set() # only store unique values 
        
        # dfs function
        def dfs(r, c, base_r, base_c, shape):
            if 0 <= r < m and 0 <= c < n and grid[i][j] == 1 and not visited[i][j]:
                visited[i][j] = True 
                shape.append((base_r - r, base_c - c))
                
                dfs(r-1, c, base_r, base_c, shape)
                dfs(r+1, c, base_r, base_c, shape)
                dfs(r, c-1, base_r, base_c, shape)
                dfs(r, c+1, base_r, base_c, shape)
        
        
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1 and not visited[i][j]:
                    shape = []
                    dfs(i, j, i, j, shape)
                    shapes.add(tuple(shape))
        return len(shapes)
```

### 12 Is Graph Bipartite 
- https://leetcode.com/problems/is-graph-bipartite/description/

```python
class Solution:
    def isBipartite(self, graph):
        V = len(graph)
        colors = [-1] * V

        for i in range(V):
            if colors[i] == -1:
                colors[i] = 0
                queue = [i]

                while queue:
                    node = queue.pop(0)

                    for neighbour in graph[node]:
                        if colors[neighbour] == -1: # Not colored
                            colors[neighbour] = 1 - colors[node]
                            queue.append(neighbour)
                        else:
                            if colors[neighbour] == colors[node]:
                                return False
        return True
```

### 13 DETECT CYCLE IN DIRECTED GRAPH
- https://www.geeksforgeeks.org/problems/detect-cycle-in-a-directed-graph/1

```python
class Solution:
    def isCyclic(self, V, edges):
        # Adjecency List 
        adj_list = [[] for _ in range(V)]
        for (u,v) in edges:
            adj_list[u].append(v)
            
        # code here
        visited = [False] * V 
        pathVisited = [False] * V 
        
        def dfs(node):
            visited[node] = True
            pathVisited[node] = True 
            
            for neighbour in adj_list[node]:
                if not visited[neighbour]:
                    if dfs(neighbour):
                        return True 
                else:
                    if pathVisited[neighbour]:
                        return True
                        
            pathVisited[node] = False
            return False
        
        for i in range(V):
            if not visited[i]:
                if dfs(i):
                    return True 
        return False
```

### 14 Topological Sorting 
- https://www.geeksforgeeks.org/problems/topological-sort/1

```python
class Solution:
    def topoSort(self, V, edges):
        # Make adjecency List  of idrected graph
        adj_list = [[] for i in range(V)]
        for (u, v) in edges:
            adj_list[u].append(v)
        stack = []
        visited = [False] * V 
        
        def dfs(node):
            visited[node] = True 
            
            for neighbour in adj_list[node]:
                if not visited[neighbour]:
                    dfs(neighbour)
            stack.append(node)
            
        
            
        for i in range(V):
            if not visited[i]:
                dfs(i)
        
        return stack[::-1]
        
```

### 15 Topological Sorting(Kahn's Algorithm) 
- https://www.geeksforgeeks.org/problems/topological-sort/1#

```python
class Solution:
    def topoSort(self, V, edges):
        #Step 1: Builf graph & indigree
        graph = [[] for _ in range(V)]
        indigree = [0] * V
        
        for (u,v) in edges:
            graph[u].append(v)
            indigree[v] += 1
        
        # Step 2: Queue for the vertices with indigree 0 
        queue = []
        for i in range(V):
            if indigree[i] == 0:
                queue.append(i)
        
        # Step 3 : process the queue 
        topo_order = []
        
        while queue:
            u = queue.pop(0)
            topo_order.append(u)
            
            for v in graph[u]:
                indigree[v] -= 1
                if indigree[v] == 0:
                    queue.append(v)
        
        # Step 4: Check the cycle 
        if len(topo_order) != V:
            return "Error"
        return topo_order
```
