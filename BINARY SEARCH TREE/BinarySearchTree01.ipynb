### 1 Search in Binary Search Tree
- https://leetcode.com/problems/search-in-a-binary-search-tree/submissions/1901443981/
- https://www.geeksforgeeks.org/problems/search-a-node-in-bst/1#

```python
'''
class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None
'''

class Solution:
    def search(self, root, key):
        # code here
        while root:
            if root.data == key:
                return True 
            if key < root.data:
                root = root.left
            else:
                root = root.right
        return False
```

### 2. ceil in Binary SearchTree
- https://www.geeksforgeeks.org/problems/implementing-ceil-in-bst/1

```python
''' class Node:
    def __init__(self, val):
        self.right = None
        self.data = val
        self.left = None 
'''
        
class Solution:
    def findCeil(self,root, x):
        # code here
        ceil = -1 
        
        while root:
            if root.data == x:
                return x 
            
            if root.data < x:
                root = root.right 
            else:
                ceil = root.data 
                root = root.left 
        return ceil
```

### 3 Floor of Binary Tree
- https://www.geeksforgeeks.org/problems/floor-in-bst/1#

```python
'''
# Node Class:
class Node:
    def _init_(self, val):
        self.data = val
        self.left = None
        self.right = None
'''

class Solution:
    def floor(self, root, x):
        # maximum(value) <= root.data 
        floor = -1 
        
        while root:
            if root.data == x:
                return root.data 
            
            if root.data > x:
                root = root.left
            else:
                floor = root.data 
                root = root.right 
        return floor
```

### 4. Insert a node in Binary Tree
- https://leetcode.com/problems/insert-into-a-binary-search-tree/submissions/1901570381/

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def insertIntoBST(self, root, val):
        if root is None:
            return TreeNode(val=val)
        
        current = root
        previous = None 
    
        while current:
            previous = current

            if current.val == val:
                return root
            
            if val < current.val:
                current = current.left
            else:
                current = current.right 
        
        # Adding the new node 
        if val < previous.val:
            previous.left = TreeNode(val=val)
        else:
            previous.right = TreeNode(val=val)
        # Return the root
        return root
```

### 5. Delete a node in a BST
- https://leetcode.com/problems/delete-node-in-a-bst/
- https://www.geeksforgeeks.org/problems/delete-a-node-from-bst/1

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def deleteNode(self, root, key):
        if root is None:
            return None
        
        # Step 1: searching the element 
        curr = root 
        parent = None

        while curr and curr.val != key:
            parent = curr

            if key < curr.val:
                curr = curr.left
            else:
                curr = curr.right 
            
        # if not element as key 
        if curr is None:
            return root
        
        # Step 2: if 2 childs
        if curr.left and curr.right:
            # Find successor
            succ_p = curr
            succ = curr.right

            while succ.left:
                succ_p = succ
                succ = succ.left
            
            # Relace the value of current
            curr.val = succ.val

            # delete the succ
            curr = succ
            parent = succ_p

        # if only 1 or 0 childs
        child = curr.left if curr.left else curr.right

        if not parent:
            return child
        
        if parent.left == curr:
            parent.left = child
        else:
            parent.right = child
        return root
```
